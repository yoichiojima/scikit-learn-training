```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
df = pd.read_csv('../data/house-prices/train.csv')
```

```python
f, ax = plt.subplots(1, 2, figsize = (8, 8))
sns.barplot(df.describe().drop(columns = 'SalePrice').T[['mean']].reset_index(), y = 'index', x = 'mean', ax = ax[0])
sns.barplot(df.describe().drop(columns = 'SalePrice').T[['std']].reset_index(), y = 'index', x = 'std', ax = ax[1])
# remove axis labels
ax[1] = ax[1].set(yticklabels = [], ylabel = '')
f.tight_layout()
plt.show()
```
